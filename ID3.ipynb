```python
import math
import pandas as pd

def calculate_entropy(labels):
    # Count how many of each class we have
    counts = {}
    for l in labels:
        counts[l] = counts.get(l, 0) + 1
    
    entropy = 0
    for c in counts.values():
        prob = c / len(labels)
        entropy -= prob * math.log2(prob)
    return entropy
```

```python
def information_gain(data, feature_name, target_name):
    # Calculate the entropy of the whole group (before split)
    total_entropy = calculate_entropy(data[target_name])
    
    # Calculate the entropy after splitting
    values = data[feature_name].unique()
    weighted_entropy = 0
    
    for val in values:
        subset = data[data[feature_name] == val]
        prob = len(subset) / len(data)
        weighted_entropy += prob * calculate_entropy(subset[target_name])
        
    # Gain = Entropy before - Entropy after
    return total_entropy - weighted_entropy
```

```python
def find_best_feature(data, target_name):
    features = data.columns.drop(target_name)
    best_feature = None
    max_gain = -1
    
    for feat in features:
        gain = information_gain(data, feat, target_name)
        print(f"Feature: {feat} | Information Gain: {gain:.4f}")
        
        if gain > max_gain:
            max_gain = gain
            best_feature = feat
            
    return best_feature
```

```python
data_dict = {
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Overcast', 'Sunny'],
    'Temp': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Mild'],
    'Play': ['No', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data_dict)

best = find_best_feature(df, 'Play')
print(f"\n--- Result: The root of your tree should be '{best}' ---")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../data/Salary_Data.csv")
df.head()
```

```python
x = df["YearsExperience"].values
y = df["Salary"].values
```

```python
plt.scatter(x,y)
```

```python
def fonction_cout(x, y, a, b):
    m = len(x)
    somme = 0
    for i in range(m):
        f = a * x[i] + b
        cout = (f - y[i]) ** 2
        somme += cout
        J = (1/(2*m)) * somme
    return J
```

```python
def gradient(x, y, a, b):
    m = len(x)
    s_a, s_b = 0, 0
    for i in range(m):
        f = a * x[i] + b
        s_a += (f - y[i]) * x[i]
        s_b += (f - y[i])
        
    dJ_da = (1/m) * s_a
    dJ_db = (1/m) * s_b
    
    return dJ_da, dJ_db
```

```python
def gradient_descent(x , y, iterations, alpha):
    a = 0 
    b = 0
    for i in range(iterations):
        dJ_da, dJ_db = gradient(x, y, a, b)

        a = a - alpha * dJ_da
        b = b - alpha * dJ_db


    return a, b
```

```python
learning_rate = 0.01
iterations = 1000

a, b = gradient_descent(x, y, iterations, learning_rate)

print(f"a: {a:.4f}, b: {b:.4f}")
```

```python
plt.scatter(x, y)
x_vals = np.linspace(min(x), max(x), 100)
y_vals = a * x_vals + b
plt.plot(x_vals, y_vals, color = 'red')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
```

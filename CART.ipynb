```python
import numpy as np
import pandas as pd

def calculate_gini(labels):
    if len(labels) == 0:
        return 0
    
    counts = {}
    for l in labels:
        counts[l] = counts.get(l, 0) + 1
        
    # Formula: 1 - Sum(P_i^2)
    sum_sq_probs = 0
    for c in counts.values():
        prob = c / len(labels)
        sum_sq_probs += prob**2
        
    return 1 - sum_sq_probs
```

```python
def find_best_binary_split(data, target_name):
    best_gini = 1.0 # Highest possible impurity
    best_feature = None
    best_value = None
    
    features = data.columns.drop(target_name)
    
    for feat in features:
        unique_values = data[feat].unique()
        
        for val in unique_values:
            # Create two groups: <= val and > val
            left_group = data[data[feat] == val][target_name]
            right_group = data[data[feat] != val][target_name]
            
            if len(left_group) == 0 or len(right_group) == 0:
                continue
                
            # Calculate weighted Gini
            w_left = len(left_group) / len(data)
            w_right = len(right_group) / len(data)
            current_gini = (w_left * calculate_gini(left_group)) + \
                          (w_right * calculate_gini(right_group))
            
            if current_gini < best_gini:
                best_gini = current_gini
                best_feature = feat
                best_value = val
                
    return best_feature, best_value, best_gini
```

```python
# Dataset: [Study Hours, Had Coffee, Passed Exam?]
study_data = {
    'Hours': [1, 2, 3, 4, 5, 6],
    'Coffee': ['No', 'No', 'Yes', 'Yes', 'No', 'Yes'],
    'Pass': ['No', 'No', 'No', 'Yes', 'Yes', 'Yes']
}

df_cart = pd.DataFrame(study_data)

feat, val, gini = find_best_binary_split(df_cart, 'Pass')

print(f"--- CART Analysis ---")
print(f"Best feature to split: {feat}")
print(f"Best split value: {val}")
print(f"Resulting Gini Impurity: {gini:.4f}")
```
